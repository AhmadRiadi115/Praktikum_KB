# Lakukan Data cleaning pada data dengan nilai null 

```python
standarisasi = df.copy()

standarisasi['Year'] = SimpleImputer(strategy='mean').fit_transform(standarisasi[['Year']])
standarisasi['Publisher'] =SimpleImputer(strategy='most_frequent').fit_transform(standarisasi[['Publisher']])

standarisasi.isna().sum()
```
