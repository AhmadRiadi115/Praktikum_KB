Import Package yg akan digunakan


```python
import numpy as np
import pandas as pd
```

Menyimpan dataset dalam sebuah variabel "Data"


```python
Data = pd.read_csv("vgsales.csv")
```

Menampilkan 10 data awal

```python
Data.head(10)
```

mencari jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe
angka, jumlah attribute yang memiliki nilai null. kemudian di tampung di dictionary

```python
jumlah = len(Data)
attribut = Data.columns
objek = len(Data.select_dtypes(include=['object']).columns)
float = len(Data.select_dtypes(include=['float64']).columns)
angka = len(Data.select_dtypes(include=['int64']).columns)
null  = len(Data.columns[Data.isna().sum() > 0])

dictionary_data = {
    "Jumlah Record      ": jumlah,
    "Nama Attribut   ": attribut,
    "Jumlah Attribut ": objek,
    "Jumlah Float    ": float,
    "Jumlah Angka    ": angka,
    "Jumlah Null     ": null,
}
```

Menampilkan isi Dictionary

```python
pd.Series(dictionary_data)
```

Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
Data.describe()
```

Menampilkan Nilai Korelasi

```python
Data.select_dtypes(include=['int64','float64']).corr()
```
