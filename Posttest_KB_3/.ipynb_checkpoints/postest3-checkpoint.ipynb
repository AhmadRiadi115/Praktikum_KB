# Import Package

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Buat Variabel untuk Menampung Dataset

```python
data = pd.read_csv('vgsales.csv')
data.head(10)
```

# Gendre beserta jumlah data

```python
data['Genre'].value_counts()
```

# Grafik Pie

```python
Gendre = ['Action',
          'Sports',
          'Misc',
          'Role-Playing',
          'Shooter',
          'Adventure',
          'Racing',                                            
          'Platform',
          'Simulation',
          'Fighting ',
          'Strategy ',
          'Puzzle']
JumlahData = data['Genre'].value_counts()

fig, ax = plt.subplots(figsize=(8,4))
ax.pie(JumlahData,startangle=90, autopct='%0.1f%%')
ax.axis('equal')
ax.legend(Gendre, title='Gendre Game', loc='lower left')
plt.ylabel('List Nama', size=14)
plt.xlabel('Presentase Game Terjual', size=14)
plt.xticks(size=8)
plt.yticks(size=12)
ax.set_title('Grafik Pie Gendre Game Terjual')

plt.show()
```

# Grafik Batang

```python
Gendre = ['Action',
          'Sports',
          'Misc',
          'Role-Playing',
          'Shooter',
          'Adventure',
          'Racing',                                            
          'Platform',
          'Simulation',
          'Fighting ',
          'Strategy ',
          'Puzzle']
JumlahData = data['Genre'].value_counts()

plt.figure(figsize=(12,7))
plt.bar(Gendre, JumlahData)

plt.title('Grafik Batang Gendre Game Terjual', size=16)
plt.ylabel('Jumlah Terjual', size=14)
plt.xlabel('Gendre', size=14)
plt.xticks(size=8)
plt.yticks(size=12)

plt.show()
```

```python
Gendre = ['Action',
          'Sports',
          'Misc',
          'Role-Playing',
          'Shooter',
          'Adventure',
          'Racing',                                            
          'Platform',
          'Simulation',
          'Fighting ',
          'Strategy ',
          'Puzzle']
JumlahData = data['Genre'].value_counts()
plt.figure(figsize=(10,5))
plt.plot(JumlahData, Gendre, label ='Garis')
plt.legend()
plt.title('Grafik Garis Gendre Game Terjual', size=16)
plt.ylabel('Gendre', size=14)
plt.xlabel('Jumlah Terjual', size=14)
plt.xticks(size=8)
plt.yticks(size=12)
plt.show()
```
